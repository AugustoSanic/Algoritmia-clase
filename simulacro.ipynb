```python
def bubble_sort_opt(arr):

    n = len(arr)

    for i in range(n):

        sorted = True

        for j in range(n - 1 - i):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                sorted = False
        
        if sorted:
            return arr
        
    return arr
```

```python
# Completar simulacro...
import random

lista_desordenada = [random.randint(0, 10_000_000) for _ in range(5000)]
print(lista_desordenada)
```

```python
# ordenar lista bubblesort
lista_ordenada_bubble = bubble_sort_opt(lista_desordenada.copy())
```

```python
def counting_sort(arr):
    # Verifica si la lista está vacía, si lo está, retorna una lista vacía
    if not arr: 
        return []

    # Obtiene la longitud de la lista y el valor máximo en la lista (k)
    n = len(arr)
    k = max(arr) + 1  # 'k' es el valor máximo más uno para ajustar el índice

    # Crea un arreglo auxiliar 'C' de tamaño 'k', inicializado en 0
    C = [0] * k

    # Recorre la lista original y cuenta la aparición de cada número en 'C'
    for i in range(n):
        C[arr[i]] += 1

    # Transforma el arreglo 'C' para que cada posición contenga la suma de las anteriores
    for i in range(1, k):
        C[i] = C[i] + C[i - 1] 

    # Crea un nuevo arreglo 'B' para almacenar la lista ordenada
    B = [None] * len(arr)

    # Recorre la lista original en orden inverso para colocar los elementos en el arreglo 'B'
    for i in range(n - 1, -1, -1):
        C[arr[i]] -= 1
        B[C[arr[i]]] = arr[i]
    
    
    return B
```

```python
# Ordenar la lista desordenada con Counting Sort
lista_ordenada_counting = counting_sort(lista_desordenada.copy())
```

```python
def buscar_y_ordenar(lista, algoritmo_ordenamiento, valor_buscado):
   
    lista_ordenada = algoritmo_ordenamiento(lista)
    
    # Búsqueda del valor en la lista ordenada
    try:
        
        return lista_ordenada.index(valor_buscado)
    except ValueError:
        # Retorna -1 si no se encuentra el valor
        return -1
```

```python
import time


inicio_bubble = time.time()
indice_bubble = buscar_y_ordenar(lista_desordenada.copy(), bubble_sort_opt, 4388485)
fin_bubble = time.time()
tiempo_bubble = fin_bubble - inicio_bubble


inicio_counting = time.time()
indice_counting = buscar_y_ordenar(lista_desordenada.copy(), counting_sort, 4388485)
fin_counting = time.time()
tiempo_counting = fin_counting - inicio_counting


print(f"Tiempo con Bubble Sort Optimizado: {tiempo_bubble} segundos")
print(f"Índice del valor en Bubble Sort: {indice_bubble}")

print(f"Tiempo con Counting Sort: {tiempo_counting} segundos")
print(f"Índice del valor en Counting Sort: {indice_counting}")
```
